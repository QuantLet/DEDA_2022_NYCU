```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

public_debts = pd.read_excel ('/Users/youcefnazimtahari/Downloads/econ-gen-poids-dette-pub-ue-2.xlsx')
print (public_debts)
```

```python
x = list(public_debts['Countries'])
y = list(public_debts['Year : 2021'])
```

```python
plt.figure(figsize=(29,10))
plt.style.use('seaborn')
plt.title("Public debts in the european union in 2021")
plt.scatter(x, y)
plt.savefig("Public debts in the european union in 2021", transparent = True)
plt.show()

```

```python
data = list(zip(x, y))

hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(data)

plt.scatter(x, y, c=labels)
plt.savefig("Cluster GDP to debt level ratio", transparent = True)

plt.show()

#the points in black represent the countries whose debt/GDP ratio is greater than 100% (Belgium, Chypre, Spain, France, Grece, Italy, Portugal)
#the points in white represent the countries whose debt/GDP ratio is greater than 100%
```
