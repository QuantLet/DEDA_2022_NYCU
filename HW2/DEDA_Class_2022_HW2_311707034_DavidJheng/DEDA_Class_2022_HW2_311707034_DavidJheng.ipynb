```python
import requests
import json
import pprint
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
response = requests.get('https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json')
content  = response.content
json_tree = json.loads(content)
df = pd.DataFrame(json_tree)
```

### 相關變數
| Variables | Meanings |
| :---: | :---: |
| tot | 總樁數 |
| sbi | 可借車輛數 |
| bemp | 可停空位數 |
| lat | 經度 |
| lng | 緯度 |
| sna | 站名 |
| sarea | 行政區 |
| sareaen | 英文行政區 |
| aren | 地址 |
| mday | 資料更新時間 |

```python
import folium

df['lat'] = pd.to_numeric(df['lat'])
df['lng'] = pd.to_numeric(df['lng'])

loc = df[['lat','lng']].values.tolist()
mapit = folium.Map( location = loc[860], zoom_start = 12.5 )
for point in range(0, len(loc)):
    information = df['sarea'][point] + ' ' + df['sna'][point][11:] + '<br>Bikes: ' + str(df['sbi'][point]) + '<br>Docks: ' +\
         str(df['bemp'][point]) + '<br>' + df['ar'][point] + '<br><br>Last Updated: ' + df['mday'][point]
    if df['sbi'][point] == 0:
        Color = 'lightgray'
    elif df['bemp'][point] == 0:
        Color = 'red'
    else:
        Color = 'beige'
    folium.Marker(loc[point], popup = folium.Popup(information, max_width=500), max_width='100', icon=folium.Icon(color=Color)).add_to(mapit)
mapit
```

```python
circle_map = folium.Map( location = loc[860], zoom_start = 12.5 )

for index, row in df.iterrows():
    information = row['sarea'] + ' ' + row['sna'][11:] + '<br>Bikes: ' + str(row['sbi']) + '<br>Docks: ' +\
         str(row['bemp']) + '<br>' + row['ar'] + '<br><br>Last Updated: ' + row['mday']
    radius = 4*row['sbi']
    folium.Circle(location=[row['lat'], row['lng']], radius=radius, popup=folium.Popup(information, max_width=500),\
         max_width='100', fill=True, fill_opacity=0.4, color='orange', weight=4.5).add_to(circle_map)
circle_map
```
